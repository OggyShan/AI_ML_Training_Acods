```python
#step 1
old_matrix=np.random.randint(0,100, size=(9,9))
print(old_matrix)

```

```python
new_matrix=np.zeros((old_matrix.shape[0],old_matrix.shape[1]))
mean_matrix=[]

zeros=0
non_zeros=0
print("Old Matrix\n",old_matrix)
for i in range(old_matrix.shape[0]):
 #step 2
  mean=old_matrix[i,:].mean()
  mean_matrix.append(mean)
#step 3
  for j in range (old_matrix.shape[1]):
    if old_matrix[i,j] < mean_matrix[i]:
      new_matrix[i,j]=old_matrix[i,j]
      zeros+=1
    else:
      new_matrix[i,j]=0;
      non_zeros+=1

print("\n", mean_matrix, "\n")
print("\nUpdated Matrix\n",new_matrix)

#step4
if (zeros+non_zeros==old_matrix.size):
  print("Updated matrix verified")
```

```python
siz=4
np.random.seed(5)
algo=np.random.randint(1,99,size=(siz,siz))
print("Input Matrix\n",algo)
result=np.zeros((int(algo.shape[0]/2),int(algo.shape[0]/2)))
indi=-1
indj=-1
for i in range(0,algo.shape[0],2):
  indi+=1
  indj-=1
  for j in range(0,algo.shape[1],2):
    indj+=1
    # indi-=1
    temp= algo[i:i+2,j:j+2]
    v=temp.max()
    result[indi,indj]=v
    print()
    # print(temp)
print("final Result")
print(result)
```

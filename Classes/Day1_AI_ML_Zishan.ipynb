```python
def numbers():
  n1=int(input("Enter number 1: "))
  n2=int(input("Enter number 2: "))
  operator=input("Enter operator (+,-,*,/): ")
  operators=["+","-","/","*"]

  while operator not in operators:
    print("Please Enter a Valid Operator and try again")
    operator=input("Enter operator (+,-,*,/): ")
  else:
    if operator=="+":
       print(n1,operator,n2,"=",n1+n2)
    elif operator=="-":
       print(n1,operator,n2,"=",n1-n2)
    elif operator=="*":
       print(n1,operator,n2,"=",n1*n2)
    elif operator=="/":
       print(n1,operator,n2,"=",n1/n2)


def play_again():
    repeat=input("Do you want to continue, \nPress y for yes, \nPress n for exit: ")
    repeats=["no","n","yes","y"]
    while repeat not in repeats:
      print("Please Enter a Valid Input and try again")
      repeat=input("Do you want to continue, \nPress y for yes, \nPress n for exit: ")
    else:
      if repeat=='y' or repeat=='yes':
        return True
      elif repeat=='n' or repeat=='no':
         return False

def game():
  numbers()
  while play_again()==True:
    game()
game()
```

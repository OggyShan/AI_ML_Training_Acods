```python
import numpy as np
import matplotlib.pyplot as plt
# create a ara of 10x10 kms and deploy the
# 50 nodes in it
num_of_sensors = 100
np.random.seed(5)
# generate the data
#data = np.random.randint(0,100,size=(num_of_sensors,2))
data = 0+(100-0)*np.random.random((num_of_sensors,2))
# display the sensors nodes
plt.figure(1)
plt.plot(data[:,0],data[:,1],'ro')
#s - square , d - diamond, o-circle,*-star
# data.shape will return num of rows and col of data
# .shape[0] will rows and .shape[1] cols

for i in range(data.shape[0]):
    dx = data[i,0]
    dy = data[i,1]
    plt.text(dx-1,dy+2,str(i+1))
```

```python
dist_matrix= np.zeros((num_of_sensors,num_of_sensors))

neigh_matrix= np.zeros((num_of_sensors,num_of_sensors))
for i in range(data.shape[0]):
  for j in range(data.shape[0]):
    dist= np.sqrt((data[i,0]-data[j,0])**2+(data[i,1]-data[j,1])**2)
    dist_matrix[i,j]=dist
dist_matrix=dist_matrix.astype(int)


ranged=30
for i in range(dist_matrix.shape[0]):
  for j in range(data.shape[0]):
    if(i!=j):
      if dist_matrix[i,j]<=ranged:
        neigh_matrix[i,j]=1


print(neigh_matrix)
neighours=[]
node_n=int(input("Enter node number: "))
plt.figure(1)
plt.plot(data[:,0],data[:,1],'ro')

for i in range(data.shape[0]):
    dx = data[i,0]
    dy = data[i,1]
    plt.text(dx-1,dy+2,str(i+1))
    # plt.plot(data[node_n,0],data[node_n,1],'bd')

for i in range(neigh_matrix.shape[0]):
  if neigh_matrix[node_n-1,i]!=0:
    neighours.append(i)
print(f"Neighours of {node_n} node are: ", neighours)

for i in range(len(neighours)):
    dxi = [data[node_n-1,0],data[neighours[i],0]]
    dyi = [data[node_n-1,1],data[neighours[i],1]]
    plt.plot(dxi,dyi,'b',linewidth=2)
    plt.plot(data[neighours[i],0],data[neighours[i],1],'gs')
plt.show()
```

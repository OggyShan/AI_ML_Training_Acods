```python
import random
intents = {
            "greeting":["Hi! how are you?","Hello what's up"],
            "goodbye":["Great day ahead","Nice to see you"],
            "address":["Address of the hotel is near Lajpat Nagar New Delhi"],
            "Rooms":["AC Rooms and Non AC Rooms"],
            "Availability":["There are 20 AC Rooms and 30 Non AC Rooms are available"],
            "Price":["AC ROOM : 10000/day ,Non AC Room: 5000/day"]
}
```

```python
def detect_response(query):
  prediction = humingbird.Text.predict(
    text = query, # inferencing text
    labels=["greeting","goodbye",
            "address","Rooms",
            "Availability","Price"] # suply potential labels
  )
  max_score=0
  for val in prediction:
    if(val['score']>max_score):
      max_score=val['score']
      max_val = val['score']
      max_val_text = val['className']
  print(max_val_text)
  return random.choice(intents[max_val_text])

n = int(input("Enter the number of iterations: "))
for i in range(n):
    query = input("Enter your query (or type 'n' to exit): ")
    if query.lower() == 'n':
        break
    response = detect_response(query)
    print("Response:", response)
```
